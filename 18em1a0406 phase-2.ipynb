### File Handling

* Performing operation on a file is called file handling
* Entering some data to a file is called as write to a file
* Retriving data from a file is called as reading from a file
* we can perform append, delete operations on files.


```python
def writetoafile(filename,data):
    with open(filename,"w") as f:
        f.write(data)
    return
filename = ("sai.txt")
writetoafile(filename,"line1\nline2\n")
```

```python
def writetoafile(filename,data):
    with open(filename,"w") as f:
        f.write(data)
    return
filename = ("krish.txt")
writetoafile(filename,"def sum1(a,b):\nc=a+b\nreturn c\nsum1(8,5)")
def readfromafile(filename):
    with open(filename,"r")as f:
        fd = f.read()
        f.close()
    return fd
readfromafile("krish.txt")
def printlines(filename):
    with open(filename,"r") as f:
        for line in f:
            print(line,end = " ")
    return
print(readfromafile(filename))
```

```python
def appendtofile(filename,data):
    with open(filename,"a") as f:
        for line in data:
            f.write("\n"+line)
        return
data=["line3","line4","line5","line6"]
filename="krish.txt"
appendtofile(filename,data)
```
